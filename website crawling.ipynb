```python
import requests ## 크롤링용
from bs4 import BeautifulSoup  ## 스크래핑용
```

```python
url = "https://www.coupang.com/np/search?component=194176&q=%EC%BB%A4%ED%94%BC&channel=user" 
```

```python
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.18363'}
```

```python
response = requests.get(url, headers = header).text
```

```python
soup_coffee = BeautifulSoup(response, "lxml")
soup_coffee
```

```python
titles = soup_coffee.select('dl.search-product-wrap div.name')  
titles
```

```python
coffee_names = [title.get_text() for title in titles] 
print(len(coffee_names))
coffee_names
```

```python
coffee_prices = soup_coffee.select('em.sale strong.price-value')  
print(len(coffee_prices))
coffee_prices
```

```python
prices = [price.get_text() for price in coffee_prices] 
prices = prices[:36]
print(len(prices))
```

```python
for k in range(len(coffee_names)): 
    print("{1} / {2}".format(k+1, coffee_names[k], prices[k])) 
```

```python
import pandas as pd
df_coffee = pd.DataFrame({'coffee name' : coffee_names, 'coffee price' : prices})
df_coffee
```

```python
df_coffee['coffee price'] = df_coffee['coffee price'].str.replace(',', '').astype('int64')
df_coffee
```

```python
## 평균 커피값 구하기
df_coffee['coffee price'].mean()
```

```python
## 가격 오름차순 정렬
df_coffee.sort_values(by='coffee price')
```
