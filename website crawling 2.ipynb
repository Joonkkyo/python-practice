```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 크롤링할 웹 사이트 정보
base_url = 'https://www.tripadvisor.co.kr'
seoul_url = '/Restaurants-g294197-Seoul.html'

# get html
response = requests.get(base_url+seoul_url)
# response.text
```

```python
# BeautifulSoup을 활용하여 데이터 파싱
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 맛집 목록 가져오기
lists = soup.findAll('div', {'class' : 'wQjYiB7z'}) 
lists
```

```python
for item in lists:
#     tem = str(item)
#     name = item.text.split('.')
#     print(name)
    print(item.text)
```

```python
# 데이터 파싱하기
rest_id_list = []
rest_name_list = []
rest_url_list = []

for item in lists:
    # 링크 가져오기
    hrefs = item.find('a')
    href = hrefs.get('href')
    full_url = base_url + href

    # 순위와 이름 분리하기, 광고 제거하기
    #[1 , 장생건강원]
    #[2,  음식점]
    #[광고음식점]
    
    name = item.text.split('.')

    if len(name) == 2:
        rest_id = name[0]
        rest_name = name[1]
        rest_url = full_url
        #데이터 프레임을 만들기 위한 리스트
        rest_id_list.append(rest_id)
        rest_name_list.append(rest_name)
        rest_url_list.append(rest_url)
#         print('[ID]' + rest_id + ' [NAME]' + rest_name + ' [URL]' + rest_url)
        
result = pd.DataFrame(data={'[ID]' : rest_id_list, '[NAME]' : rest_name_list, '[URL]' : rest_url_list})
# print(result)
# result
```
